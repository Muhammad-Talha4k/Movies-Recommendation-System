```python
# Movie Recommendation System using Content Based Filtering

# importing the libraries

# to perform a wide variety of mathematical operations on list/arrays
import numpy as np

# Reading the CSV files
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
#Convert a collection of raw data to a matrix of TF-IDF features
from sklearn.feature_extraction.text import TfidfVectorizer 

from sklearn.metrics.pairwise import sigmoid_kernel

# Read the credits csv file and store data
credits = pd.read_csv("D:\credits.csv")

# Show the first 6 rows of data
credits.head(3) 
```

```python
# Read the movies csv file and store data
movies = pd.read_csv("D:\movies.csv")

# Show the first 6 rows of data
movies.head(6)
```

```python
# Merging the credits & movies csv files

final = pd.merge(credits,movies)

final
```

```python
# choosing important features which helps to build a movie recommendation system
final=final[["overview","original_title","id","genres","original_language","vote_average"]]

# Display the sum of all null values 
final.isnull().sum()

# Delete all null values 
final.dropna(inplace=True)

# Counts the duplicate values 
final.duplicated().value_counts()

# Delete the duplicate values 
final=final.drop_duplicates(keep="first",inplace=False)
```

```python
# Helps model to group similar words, removes all the unnecessary things from the overview column

tfv=TfidfVectorizer(min_df=3,max_features =None, strip_accents= "unicode",analyzer='word',token_pattern=r'\w{1,}',
                    ngram_range=(1,3),stop_words='english')

# converting the overview column data to matrix form
tfv_matrix=tfv.fit_transform(final["genres"],["vote_average"])

# To get the current shape of the list
tfv_matrix.shape

sig=sigmoid_kernel(tfv_matrix,tfv_matrix)


sig

index1=pd.Series(final.index,index=final["original_title"]).drop_duplicates()

index1

def give_recommendation(title,sig=sig):
    idx=index1[title]
    sigmoid_score= list(enumerate(sig[idx]))
    sigmoid_score=sorted(sigmoid_score, key=lambda x:x[1],reverse=True)
    sigmoid_score=sigmoid_score[1:6]
    movie_indices=[i[0] for i in sigmoid_score]
    return final["original_title"].iloc[movie_indices]
```

```python
#      Some movies names from the csv file :

# Batman v Superman: Dawn of Justice
# Captain America: Civil War
# Furious 7
# #Spider-Man 2
# #Man of Steel
# Superman Returns
# The Dark Knight Rises                               
# Won't Back Down
# U2 3D
# 

give_recommendation("Man of Steel")
```
